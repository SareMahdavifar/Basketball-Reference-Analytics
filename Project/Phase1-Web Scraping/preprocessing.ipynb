```python
import pandas as pd
```

```python
df = pd.read_csv('./player_stats.csv')

stats_df = df.copy()
stats_df = stats_df[stats_df["season"] != "2024-25"]

for c in stats_df.columns:
    try:
        stats_df[c] = pd.to_numeric(stats_df[c])
    except :
        pass

stats_df
```

```python
df_copy = stats_df[stats_df['Awards'].notna() & (~stats_df['Team'].str.match(r'^\d+TM$'))]

awards_df = df_copy[['season', 'Player', 'Team', 'Awards']].copy()
awards_df['Awards'] = awards_df['Awards'].str.split(',')
awards_df = awards_df.explode('Awards')

awards_df
```

```python
def seperate_mvp_rank(awards):
    items = str(awards).split(',')
    for item in items:
        if item.startswith('MVP-'):
            return int(item.split('MVP-')[1])
    return 0

stats_df['mvp_rank'] = stats_df['Awards'].apply(seperate_mvp_rank)

stats_df.drop(columns='Awards', inplace=True)

stats_df
```

```python
total_nulls = stats_df.isna().sum().sum()
print(f'Total NULL: {total_nulls}\n')

nulls = stats_df.isna().sum()
print(f'Null in each column:\n{nulls[nulls > 0]}\n')
```

```python
stats_df = stats_df.fillna(0)

stats_df.isna().sum().sum()
```

```python
def convert_season(season_year):
    separator = '_' if '_' in season_year else '-'
    first, second = season_year.split(separator)
    return f'{first}-{int(second)-2000}'
```

```python
# prepare team_seasons table

team_seasons_df = pd.read_excel('Teams_Final.xlsx', sheet_name='Teams_Final')

team_seasons_df['season'] = team_seasons_df['season'].apply(convert_season)
team_seasons_df.drop(columns='team_link', inplace=True)

team_seasons_df['season'] = team_seasons_df['season'].map(season_map)
team_seasons_df.rename(columns={'season': 'season_id'}, inplace=True)

team_seasons_df['team_name'] = team_seasons_df['team_name'].map(team_map)
team_seasons_df.rename(columns={'team_name': 'team_id'}, inplace=True)

team_seasons_df['id'] = range(1, len(team_seasons_df) + 1)

with pd.ExcelWriter('player_stats.xlsx', engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    team_seasons_df.to_excel(writer, sheet_name='team_seasons', index=False)

team_seasons_df
```

```python
# prepare stats table

stats_df = pd.read_excel('player_stats.xlsx', sheet_name='stats')

# # add salary
# salary_df  = pd.read_csv('./salary.csv')

# salary_df['season_name'] = salary_df['season_name'].apply(convert_season)

# merge_salary = stats_df.merge(salary_df[['player_name', 'season_name', 'team_name', 'salary']], on=['player_name', 'season_name', 'team_name'], how='left')

# with pd.ExcelWriter('player_stats.xlsx', engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
#     merge_salary.to_excel(writer, sheet_name='check', index=False)
# merge_salary


# add experience

roster_df = pd.read_excel('Roster.xlsx', sheet_name='stats')
roster_df.rename(columns={'season': 'season_name'}, inplace=True)

roster_df['season_name'] = roster_df['season_name'].apply(convert_season)

merge_experience = stats_df.merge(roster_df[['player_name', 'season_name', 'team_name', 'experience']], on=['player_name', 'season_name', 'team_name'], how='left')

portable_players = merge_experience['team_name'].str.contains('TM', na=False)

first_team = merge_experience['team_name'].where(~portable_players).groupby([merge_experience['player_name'], merge_experience['season_name']]).transform('first')
merge_experience.loc[portable_players, 'team_name'] = first_team

merge_experience.loc[portable_players, 'experience'] = merge_experience.groupby(['player_name', 'season_name'])['experience'].transform('first')

merge_experience = merge_experience[~merge_experience.duplicated(['player_name', 'season_name'], keep='first')]

merge_experience['innate_ability'] = round(merge_experience['experience']/ merge_experience['age'],2)

merge_experience.drop('player_url', axis=1, inplace=True)

merge_experience['season_name'] = merge_experience['season_name'].map(season_map)
merge_experience['position'] = merge_experience['position'].map(position_map)
merge_experience['player_name'] = merge_experience['player_name'].map(players_map)
merge_experience['team_name'] = merge_experience['team_name'].map(team_map)

merge_experience['multi_team_flag'] = 0
merge_experience.loc[portable_players, 'multi_team_flag'] = 1

merge_experience.rename(columns={'season_name': 'season_id', 'position': 'position_id', 'player_name': 'player_id', 'team_name': 'team_id'}, inplace=True)

merge_experience['id'] = range(1, len(merge_experience) + 1)

with pd.ExcelWriter('player_stats.xlsx', engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    merge_experience.to_excel(writer, sheet_name='stats', index=False)
merge_experience
```
